```python
import csv
with open('california_housing_test.csv') as file_obj:
  file_data=csv.reader(file_obj,delimiter=",",skipinitialspace=True)
  # delimiter is used to seperate the data if there is a other seperator than comma
  # skipinitial spaces will remove the spces before

  for row in file_data:
    print(row)
```

```python
import csv
with open('california_housing_test.csv') as file_obj:
  file_data=csv.reader(file_obj)

  file_list=list(file_data)
  # this command is used to put all the seperate line of code to a single 2d list so that we can traverse upon each
  # print(file_list)
  total_rooms=0
  for i in file_list[1:]:
    # print(i[5])
    total_rooms+=int(float(i[5]))
  
  print(total_rooms)

 
```

```python
import csv
with open('california_housing_test.csv') as file_obj:
  file_data=csv.DictReader(file_obj)
  # DictReader will store the data in key value pairs
  for i in file_data:
    print(i)
    
```

```python
import csv
with open('california_housing_test.csv') as file_obj:
  file_data=csv.DictReader(file_obj,fieldnames={'a','b','c','d','e','f'})
  # we can also create our own field names 
  # if we did not give our field names for any column so they will be asigned to null
  # it is very useful if there is no fieldanames
  

  for i in file_data:
    print(i)
```
